```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, label='Boston', color="blue", linewidth=2)

plt.xlabel("Date")
plt.ylabel("Fare($)")

plt.ylim(0, 45)

plt.title("Pyber Fare By Month")
plt.grid()
plt.legend()


```

```python
fig, ax = plt.subplots()

ax.plot(x_axis, y_axis, color='green', marker='D', linewidth=2, label='Boston')
ax.grid()
ax.legend()

ax.set_title("Pyber Fare By Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

ax.set_ylim(0,45)
```

```python
plt.bar(x_axis, y_axis, color='green', label='Boston')

plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")

plt.legend()
```

```python
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()

plt.title("PyBer Fare By Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')

fig.gca().invert_yaxis()

ax.set_xlabel("Fare")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare By Month")

ax.legend()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c='red', label='Chicago')

plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare By Month")

plt.xlim(0,50)
plt.gca().invert_yaxis()

plt.legend()
```

```python
plt.scatter(y_axis, x_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue", s =[i * 5 for i in y_axis], label="Boston", alpha=0.8, edgecolor="black", linewidth=2)

ax.set_title("PyBer Fare By Month")
ax.set_ylabel("Fare($)")
ax.set_xlabel("Date")

fig.gca().invert_yaxis()

ax.set_xlim(0, 50)

ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
        explode=explode_values,
        labels=x_axis,
        autopct='%.1f%%',
        colors=colors)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "skyblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis,
       labels=x_axis,
       autopct='%.1f%%',
       explode=explode_values,
       shadow =True,
       startangle =85,
       colors=colors,)
plt.show()
```
